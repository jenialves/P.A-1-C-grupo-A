Questão 6 – Cálculo de salário com bônus

Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras: • Pontuação menor que 5: sem bônus • Pontuação de 5 a 7: bônus de 5% do salário • Pontuação de 8 a 9: bônus de 10% do salário • Pontuação igual a 10: bônus de 15% do salário

```python
#questão 1

idade = int(input("Digite a sua idade: "))


if idade <= 17:
    print("Você é menor de idade.")

if (idade >= 18) and (idade <= 59):
    print("Você é Adulto.")

if idade >= 60:
    print("Você é Idoso.")
```

```python
#questão 2

valor = float(input("digite o valor total da compra: "))


if (valor >= 101) and (valor <= 500):
    calc = valor * 0.10
    total = valor - calc
    print(f"O valor total da sua compra com 10% de desconto é de: {total}")

if valor > 500:
    calc = valor * 0.20
    total = valor - calc
    print(f"O valor total da sua compra com 20% de desconto é de: {total}")

if valor <= 100:
    print(f"O valor total da sua compra é de: {valor}")
```

```python
# questão 3

print('Validador de triângulos')
lado1 = float(input("Digite o valor do primeiro lado: "))
lado2 = float(input("Digite o valor do segundo lado: "))
lado3 = float(input("Digite o valor do terceiro lado: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os valores podem formar um triângulo válido.")
else:
    print("Os valores NÃO podem formar um triângulo válido.")
```

```python
# questão 4

energia = int(input("Digite a energia do seu personagem: "))
print(f"{energia} de energia")

if (energia >= 71) and (energia <= 100):
    print(f"O seu personagem ainda está com {energia} de energia, ele está saudável")

if (energia >= 31) and (energia <= 70):
    print(f"O seu personagem ainda está com {energia} de energia, ele está Normal")

if (energia >= 1) and (energia <= 30):
    print(f"O seu personagem ainda está com {energia} de energia, ele está em estado Critico")

if energia == 0:
    print(f"Você está com {energia} de energia, você está Morto")
```
