```python
#QUESTÃO 02 – Soma dos Números (2 pontos)

#Solicite ao usuário um número inteiro positivo.
#Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.

numero = int(input("digite um numero: "))
soma = 0
for i in range(1, numero + 1):
  soma += i
print(f"A soma até {numero} é: {soma}")
```

```python
#QUESTÃO 03 – Tabuada (2 pontos)

#Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.
# escrever um programa que o usuário digita um número de 1 a 10 e imprime a tabuada.

número = int (input("digite um número: "))
for i in range (0, 11):
  print (f"{número} x {i} = {número * i}")
```

```python
#QUESTÃO 04 – Fatorial (4 pontos)

#Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.
#Observação: O programa deve validar se o número informado é positivo.


import math
numero = int (input("digite o número desejado: "))
while numero < 0:
  numero = int (input("por favor, digite um número inteiro e positivo: "))
fatorial = math.factorial(numero)
print(f"A fatoração de {numero} é = {fatorial}")
```

```python
# QUESTÃO 05 – Caixa Eletrônico (5 pontos)

#Um caixa eletrônico possui notas de: R$ 100, R$ 50, R$ 20, R$ 10, R$ 5, R$ 2.

#Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

#Requisitos:
#- Utilizar laços de repetição.
#- Não utilizar listas.
#- Exibir apenas as notas utilizadas.

valor_saque = int(input("Digite o valor do saque: "))
total = valor_saque

ced = 100
while True:
    totalced = 0
    while total >= ced:
        total -= ced
        totalced += 1
    if totalced > 0:
        print(f"{totalced} nota(s) de R$ {ced}")

    if total == 0:
        break


    if ced == 100:
        ced = 50
    elif ced == 50:
        ced = 20
    elif ced == 20:
        ced = 10
    elif ced == 10:
        ced = 5
    elif ced == 5:
        ced = 2
    elif ced == 2:
        ced = 1
```
