operadores aritméticos: soma de dois números

```python
#entrada

a = int (input("digite o primeiro número: "))
b = int (input("digite o segundo número: "))

#processamento

soma = a + b
subtração = a - b
multiplicação = a * b
divisão = a / b
div_int = a // b
resto = a % b #resto da divisão

#saída

print ("")
print ("------------SOMA--------------")
print (f"a soma de {a} mais {b} é igual a: {soma} ")
print ("")
print ("------------SUBTRAÇÃO--------------")
print ("")
print (f"a diferença de {a} menos {b} é igual a: {subtração} ")
print ("")
print ("------------MULTIPlICAÇÃO--------------")
print (f"a multiplica de {a} vezes {b} é igual a: {multiplicação} ")
print ("")
print ("------------DIVISÃO--------------")
print ("")
print (f"a divisão de {a} dividido {b} é igual a: {divisão} ")
print ("")
print ("------------DIVISÃO INTEIRA--------------")
print ("")
print (f"a divisão de {a} dividido {b} é igual a: {div_int} ")
print ("")
print ("------------RESTO--------------")
print ("")
print (f"a divisão de {a} dividido {b} é igual a: {resto} ")
print ("-----------------------------------")
```

operadores relacionais

```python
#entrada

num1 = int (input("digite um número: "))
num2 = int (input("digite um número: "))

#processamento

igual = num1 == num2
diferente = num1 != num2
maior =  num1 > num2
menor = num1 < num2
menorigual = num1 <= num2
maiorigual = num1 >= num2

#saída

print (igual)
print (diferente)
print (maior)
print (menor)
print (menorigual)
print (maiorigual)
```
