programa que o usuário digita seu nome e idade e o programa retorna se ele é criança, adolescente ou adulto.

```python
#entrada

nome = str (input("digite o seu nome: "))
idade = int (input("digite sua idade: "))

#processamento e saída

if idade < 12:
  print(f"{nome}, você é criança.")

elif idade < 18:
   print(f"{nome}, você é adolescente.")

else:
  print(f"{nome}, você é adulto.")
```

programa que classifica os triângulos

verifica se forma um triângulo:

isóceles:

escaleno:

equilátero:

```python
lado1 = int (input("digite o tamanho do lado: "))
lado2 = int (input("digite o tamanho do lado: "))
lado3 = int (input("digite o tamanho do lado: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print (" o triângulo é válido")

if lado1 == lado2 == lado3:
 print (f"o triângulo é equilátero")

elif lado1 != lado2 or lado1 != lado3 or lado2 != lado3:
  print (f"o triângulo é escaleno")

else:
  print (f"o triângulo é isóceles")
```
