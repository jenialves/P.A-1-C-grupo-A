```python
#10 Crie um jogo de adivinhaçao, o usuario tenta adivinhar um numero secreto
#informar se o numero secreto e maior ou menor
# Se o usuario acertar, exibir uma mensagem de "você acertou!"

import random

secret_number = random.randint(1, 10) # Let's make the range a bit larger for a proper game
print("Eu pensei em um número entre 1 e 10. Tente adivinhar!")

guess = 0
while guess != secret_number:
  try:
    guess = int(input("digite seu palpite: "))
    if guess < secret_number:
      print("O número secreto é maior!")
    elif guess > secret_number:
      print("O número secreto é menor!")
    else:
      print(f"Você acertou! O número secreto era {secret_number}!")
  except ValueError:
    print("Por favor, digite um número válido.")
```
