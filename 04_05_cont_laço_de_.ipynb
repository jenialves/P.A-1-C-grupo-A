```python
#peça um número e mostre todos os númros pares até esse número

numero = int(input("digite um numero: "))
for i in range (numero):

  if i % 2 == 0:
   print (i)
```

```python
# peça um número e calcule a soma de 1 até o número
numero = int(input("digite um numero: "))
soma = 0
for i in range(1, numero + 1):
  soma += i
print(f"A soma de 1 até {numero} é: {soma}")
```

```python
#peça uma palavra e conte as letras. ex: Phyton - a palavra tem 6 letras
palavra = str(input("digite uma palavra: "))
numero_de_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_de_letras} letras.")
```

```python
# peça uma palavra e conte as vogais
palavra = str(input("digite uma palavra: "))

vogais_encontradas = 0
vogais = "aeiouAEIOU"

for letra in palavra:
  if letra in vogais:
    vogais_encontradas += 1

print(f"A palavra '{palavra}' tem {vogais_encontradas} vogais.")
```

```python
# peça para o usuario digitar 5 numeros e informe qual e o maior
numero1 = int(input("digite o primeiro numero: "))
numero2 = int(input("digite o segundo numero: "))
numero3 = int(input("digite o terceiro numero: "))
numero4 = int(input("digite o quarto numero: "))
numero5 = int(input("digite o quinto numero: "))

maior_numero = max(numero1, numero2, numero3, numero4, numero5)
print(f"O maior numero é: {maior_numero}")
```
