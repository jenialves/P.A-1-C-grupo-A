escreva um programa que verifica se o número é par ou ímpar

```python
número = int (input("Digite um numero: "))

divisao = numero % 2
if divisao == 0:
  print(f"o número {número} é par")

else:
  print(f"o número {número} é impar")
```

escrever um programa que fala se o aluno foi aprovado, está em recuperação ou foi reprovado

```python
nota1 = float (input("digite a nota do 1° bimestre: "))
nota2 = float (input("digite a nota do 2° bimestre: "))
nota3 = float (input("digite a nota do 3° bimestre: "))
nota4 = float (input("digite a nota do 4° bimestre: "))

média = (nota1 + nota2 + nota3 + nota4)/4

if média >= 7:
  print (f"você foi aprovado, sua média foi: {média} ")

elif média >= 5:
  print(f"você está de recuperação, sua média foi: {média}")

else:
  print(f"você foi reprovado, sua média foi: {média}")
```
