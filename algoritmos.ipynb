escrever um algoritmo e o programa que recebe 04 notas do aluno e calcula a média

```python
# entrada

nota_1 = 7
nota_2 = 6
nota_3 = 9
nota_4 = 8

# processamento

média = (nota_1 + nota_2 + nota_3 + nota_4)/4

# saída

print (f"média final: {média}")  #pode ser print(média)
```
